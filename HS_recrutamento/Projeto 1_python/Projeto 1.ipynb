SLOT MACHINE
1. Quando iniciada deve perguntar ao utilizador quantos créditos quer depositar
2. Antes de cada rodada deve perguntar ao utilizador se quer parar de jogar
3. Se a decisão for continuar deve-se perguntar de novo quantos créditos vão ser apostados (se não houver créditos    
   suficientes o jogador é informado e repete-se a pergunta)
4. Uma rodada consiste em gerar 3 símbolos de um conjunto de 7 símbolos (à escolha)
5. Os simbolos devem ter as seguintes probabilidades 50/156, 40/156, 30/156, 20/156, 10/156, 5/156, 1/156.
6. Se os três símbolos forem iguais então o utilizador ganha créditos em função da sua aposta e do símbolo: 
   5 * aposta símbolo 1, 10 * aposta símbolo 2, 20 * aposta símbolo 3, 70 * aposta símbolo 4, 200 * aposta símbolo 5, 
   1000 * aposta símbolo 6, 100000 * aposta símbolo 7.
7. Se o utilizador ficar sem créditos o programa deve fechar.

```python
from random import randint 
import numpy as np

cred=int(input('Quantos créditos deseja depositar? '))

probsimb=np.ones(156)
for i in range(156):
    if i>49 and i<90:
        probsimb[i]=2
    elif i>89 and i<120:
        probsimb[i]=3
    elif i>119 and i<140:
        probsimb[i]=4
    elif i>139 and i<150:
        probsimb[i]=5
    elif i>149 and i<155:
        probsimb[i]=6
    elif i>154:
        probsimb[i]=7

class Simbolo:
    simbc=1
    def roll(self):
        self.simbc=probsimb[randint(0,155)]
    def gain(self,user):
        if self.simbc==1:
            return 5*user
        if self.simbc==2:
            return 10*user
        if self.simbc==3:
            return 20*user
        if self.simbc==4:
            return 70*user
        if self.simbc==5:
            return 200*user
        if self.simbc==6:
            return 1000*user
        if self.simbc==7:
            return 100000*user
cont="sim"
while cont=="sim":
    if cred!=0:
        
        print("********************************************************************************")
        
        print("Possui", cred, "créditos")  
        
        print("********************************************************************************")
            
        user=int(input("Quantos créditos deseja apostar? "))
        while user>cred or user<=0:
            user=int(input("Créditos insuficientes. Quantos créditos deseja apostar? "))

        cred=cred-user    


        first_simbolo=Simbolo()
        first_simbolo.roll()
        second_simbolo=Simbolo()
        second_simbolo.roll()
        third_simbolo=Simbolo()
        third_simbolo.roll()
        print("********************************************************************************")
        print(int(first_simbolo.simbc), int(second_simbolo.simbc), int(third_simbolo.simbc))

        if first_simbolo.simbc==second_simbolo.simbc==third_simbolo.simbc:
            print("Ganhou", third_simbolo.gain(user))
            cred=cred+third_simbolo.gain(user)
        else:
            print("ups")
        print("********************************************************************************")
        cont=input("Deseja continuar? (insira sim ou não) ")

    else:
        print("Não possui mais créditos")
        break 
print("Acabou com", cred, "créditos")
```
